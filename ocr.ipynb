```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img_path='textsize.png'
reader = easyocr.Reader(['en'],gpu=True)
result = reader.readtext(img_path)
for i in result:
    print('position:',i[0],'|| Text:',i[1],'|| conf:',i[2])
```

```python
img_path='cr7.jpg'
reader = easyocr.Reader(['en'],gpu=True)
result = reader.readtext(img_path)
top_left=result[0][0][0]
bottom_right=result[0][0][2]
text=result[0][1]
font=cv2.FONT_HERSHEY_SIMPLEX
img=cv2.imread(img_path)
img=cv2.rectangle(img,top_left,bottom_right,(0,255,0),5)
img=cv2.putText(img,text,top_left,font,2,(255,0,0),10)
plt.imshow(img)
plt.show()
```

```python
img_path='cristiano.jpeg'
reader = easyocr.Reader(['en'],gpu=True)
result = reader.readtext(img_path)
spacer = 100
font=cv2.FONT_HERSHEY_COMPLEX
img=cv2.imread(img_path)
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),1)
    img = cv2.putText(img,text,(0,spacer), font,0.5,(0,255,0),1,cv2.LINE_AA)
    spacer+=15
    
plt.imshow(img)
plt.show()
```
